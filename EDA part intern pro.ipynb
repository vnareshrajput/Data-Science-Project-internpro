```
import matplotlib.pyplot as plt
import seaborn as sns
```


## 1. Univariate Analysis

### âž¤ What We Did:
We explored the distribution of individual variables using **histograms**, **KDE plots**, and **boxplots**.

### âž¤ Key Indicators Analyzed:
- `health_exp` - Health expenditure
- `life_expect` - Life expectancy
- `maternal_mortality`
- `infant_mortality`
- `under_5_mortality`
- `hiv_risk_score`
- `disease_burden_score`

### âž¤ Insights:
- **Boxplots** helped detect potential outliers in health-related metrics.
- Distributions showed skewness in mortality and expenditure metrics.


```
sns.set(style="whitegrid")

# List of columns for histograms/KDE/Boxplots
cols = ['health_exp', 'life_expect', 'maternal_mortality',
        'infant_mortality', 'under_5_mortality',
        'hiv_risk_score', 'disease_burden_score']

# Histograms & KDE
for col in cols:
    plt.figure(figsize=(8, 5))
    sns.histplot(df[col], kde=True, bins=30, color='skyblue')
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.tight_layout()
    plt.show()
```

```
# Boxplots for outlier detection
for col in cols:
    plt.figure(figsize=(6, 4))
    sns.boxplot(data=df, x=col, color='salmon')
    plt.title(f'Boxplot of {col}')
    plt.tight_layout()
    plt.show()
```

## 2. Bivariate Analysis

### âž¤ What We Did:
Analyzed relationships between two variables using **scatter plots**, a **correlation heatmap**, and **grouped bar plots**.

### âž¤ Key Comparisons:
- `health_exp` vs `life_expect`
- `disease_burden_score` vs `life_expect`
- `maternal_mortality` vs `infant_mortality`

### âž¤ Heatmap:
Revealed correlations between multiple numerical features, helping us understand which indicators move together.

### âž¤ Developed vs Developing Comparison:
Used bar plots to compare average indicator values based on the `is_developed` binary variable.

### âž¤ Insights:
- Higher health expenditure is generally associated with higher life expectancy.
- Disease burden and mortality are inversely related to life expectancy.


```
# Scatter: health_exp vs life_expect
plt.figure(figsize=(6, 4))
sns.scatterplot(data=df, x='health_exp', y='life_expect', hue='is_developed', palette='coolwarm')
plt.title('Health Expenditure vs Life Expectancy')
plt.tight_layout()
plt.show()
```

```
plt.figure(figsize=(6, 4))
sns.scatterplot(data=df, x='disease_burden_score', y='life_expect', hue='is_developed', palette='viridis')
plt.title('Disease Burden vs Life Expectancy')
plt.tight_layout()
plt.show()
```

```
# Scatter: maternal_mortality vs infant_mortality
plt.figure(figsize=(6, 4))
sns.scatterplot(data=df, x='maternal_mortality', y='infant_mortality')
plt.title('Maternal vs Infant Mortality')
plt.tight_layout()
plt.show()
```

```
# Correlation Heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(df[cols].corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap of Numerical Features')
plt.tight_layout()
plt.show()
```

## 3. Multivariate Analysis

### âž¤ What We Did:
Used **pairplots** to observe relationships among four important health features simultaneously.

### âž¤ Grouped Bar Plots:
Visualized multiple indicators:
- By **country** (top 20/30 based on life expectancy)
- By **development status**
- Horizontal bar chart for better readability

### âž¤ Insights:
- Countries with high life expectancy tend to invest more in health.
- Developing countries show higher under-5 and maternal mortality.


```
# Bar plot: is_developed vs average values
avg_values = df.groupby('is_developed')[cols].mean().T
avg_values.plot(kind='bar', figsize=(12, 6))
plt.title('Average Health Indicators: Developed vs Developing')
plt.ylabel('Average Value')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```
# Pairplot
sns.pairplot(df[['life_expect', 'health_exp', 'under_5_mortality', 'prev_undernourishment']], corner=True)
plt.suptitle("Pairplot of Selected Features", y=1.02)
plt.tight_layout()
plt.show()
```

```
# Select top 20 countries by life expectancy
top_countries = df.sort_values(by='life_expect', ascending=False).head(20)

indicators = ['health_exp', 'life_expect', 'under_5_mortality']
df_melted = top_countries.melt(id_vars=['country'], value_vars=indicators)

plt.figure(figsize=(14, 6))
sns.barplot(data=df_melted, x='country', y='value', hue='variable')
plt.title('Top 20 Countries by Life Expectancy - Health Indicators')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```
indicators = ['health_exp', 'life_expect', 'under_5_mortality']
grouped = df.groupby('is_developed')[indicators].mean().reset_index()
grouped['is_developed'] = grouped['is_developed'].map({0: 'Developing', 1: 'Developed'})

df_melted = grouped.melt(id_vars='is_developed', value_vars=indicators)

plt.figure(figsize=(8, 5))
sns.barplot(data=df_melted, x='is_developed', y='value', hue='variable')
plt.title('Average Health Indicators: Developed vs Developing Countries')
plt.ylabel('Average Value')
plt.tight_layout()
plt.show()
```

## ***5 . Comparative Country-Level Analysis***
Top 10 countries with highest/lowest:

### *Life* *Expectancy*

Top 10: Dominated by developed nations with strong healthcare systems (e.g., Japan, Switzerland).

Bottom 10: Primarily developing countries facing poverty, conflict, or limited healthcare access (e.g., Sierra Leone, Chad).

### *Maternal* *Mortality*

Top 10 (highest rates): Concentrated in Sub-Saharan Africa and South Asia, reflecting poor access to maternal care.

Bottom 10 (lowest rates): Found in wealthy, developed nations with robust healthcare infrastructure.

### Health Efficiency Ratio

Top 10: Includes countries that achieve good health outcomes with moderate spending (e.g., Singapore, South Korea).

Bottom 10: Either underspend or overspend inefficiently, leading to poor outcomes relative to cost.

```
latest_year = df["year"].max()
recent_df = df[df["year"] == latest_year]

def get_top_bottom(df, column, top_n=10):
    top = df.nlargest(top_n, column)[["country", column]]
    bottom = df.nsmallest(top_n, column)[["country", column]]
    return top, bottom

metrics = ["life_expect", "maternal_mortality", "health_efficiency_ratio"]

fig, axs = plt.subplots(len(metrics), 2, figsize=(16, 18))
for i, metric in enumerate(metrics):
    top, bottom = get_top_bottom(recent_df, metric)
    sns.barplot(x=metric, y="country", data=top, ax=axs[i][0], palette="Greens_d")
    axs[i][0].set_title(f"Top 10 Countries by {metric.replace('_', ' ').title()}")
    sns.barplot(x=metric, y="country", data=bottom, ax=axs[i][1], palette="Reds_d")
    axs[i][1].set_title(f"Bottom 10 Countries by {metric.replace('_', ' ').title()}")
plt.tight_layout()
plt.show()
```

# **6 . Distribution by Development Status**


 Health Metrics by Development Status

-***Life*** ***Expectancy***: Higher and more consistent in developed countries.

-**Maternal** & **Child** **Mortality**: Much lower in developed countries; developing countries show wide variability.

-**Health** **Expenditure**: Developed nations spend significantly more on healthcare.

-**Health** **Efficiency**: Generally higher in developed regions, reflecting better outcomes per dollar spent.


```
health_metrics = [
    "life_expect", "maternal_mortality", "infant_mortality", "neonatal_mortality",
    "under_5_mortality", "health_exp", "health_efficiency_ratio"
]

fig, axs = plt.subplots(len(health_metrics), 1, figsize=(12, 30))
for i, metric in enumerate(health_metrics):
    sns.violinplot(data=df, x="is_developed", y=metric, ax=axs[i], palette="Set2")
    axs[i].set_title(f"{metric.replace('_', ' ').title()} by Development Status")
    axs[i].set_xlabel("Development Status (0 = Developing, 1 = Developed)")
plt.tight_layout()
plt.show()
```
